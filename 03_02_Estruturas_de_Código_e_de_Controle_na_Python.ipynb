# Estruturas de Código e Controle na Python

Na maioria das linguagens, assim como na Python o programa deve ser capaz de tomar decisões com base em valores e resultados gerados durante sua execução, ou seja, deve ser capaz de decidir se determinada instrução deve ou não ser executada de acordo com uma condição.
Para atender a esse tipo de situação, podemos utilizar instruções especiais denominadas estruturas condicionais.

Então vamos começar com a teoria da estrutura de condições, como seu nome já diz, a condição verifica os argumentos passados e, executa um comando caso a condição seja verdadeira, como se pode ver no algorítimo abaixo:










````
SE condição
  ENTÃO comando
````

Nessa etapa vamos conversar sobre estrutura condicional. As estruturas condicionais é onde nós começamos a adicionar mais inteligência. A execução sequencial, você apenas dá o comando e a python segue, ela faz.
Mas execução sequencial é onde você pode escolher um dos dois caminhos. Vire à esquerda, vire à direita ou então siga em frente. Você pode fazer escolhas e está é uma parte essencial de fazer com que os computadores pareçam mais inteligentes.


---

Então a chave para isso é a declaração IF.
Essa palavra é reservada para indicar que vamos fazer algo condicionalmente e não é tão difícil de ler.

Para isso a declaração IF tem uma condição, que é realmente uma pergunta.

````
# Passos para a condicional
x = 5
if x < 10: 
  print("smaller")
if x > 20:
  print("Bigger")
print("Finis")
````

# Estruturas de controle

Daqui em diante, vamos observar de fato as estruturas

## Alavanca "If"

Como já mostramos, o "If", ou "Se", testa a verdade de uma condição, e executa um processo caso a condição seja verdade.

````
## Se x for maior que 2, ele irá exibir verdade!
if x > 2:
  print('É verdade!')
````

Lembre-se que é necessário antes pré-determinar o x!

````
## Determinando o x:
x = 3

## Agora, testamos de fato.
if x > 2:
  print('É verdade!')
````

- Algumas outras estruturas de dados podem ser comparadas com operadores booleanos?
  - Veja como funciona com caracteres/strings:

````
## Primeiro, vamos comprar uma batata.
comprei = "batata"

if comprei == "batata":
  print("Hoje teremos lanche!")
````

````
## Supondo algumas listas de convidados
lista1 = ["alfredo", "carlinha", "marcos"]
lista2 = ["alfredo", "marcos", "carlinha"]
listaConvidados = ["alfredo", "carlinha", "marcos"]

## Podemos comparar também?
if lista1 == lista2:
  print("A lista 2 contém os convidados!")

## A ordem precisa ser exata?
if lista1 == listaConvidados:
  print("A lista na ordem certa.")
````

Com isso, entendemos que listas podem ser comparadas usando "If" e booleanos, contanto que a lista esteja exatamente na mesma ordem.

Algumas linguagens (como Shellscript) não permitem os operadores booleanos em sua forma algébrica para comparar nada exceto numerais.


- É possível adicionar muitas operações dentro de um mesmo if:


````
## x denovo...
x = 5

## Se x for maior que 2, exaustivamente me diga que ele é.
if x > 2:
  print("É maior que 2.")
  print("Continua sendo maior que 2.")
  print("Já conversamos sobre isso. x é maior que 2.")
````

- Ainda é possível, colocar vários "If" um dentro do outro:

````
## Começando com um x
x = 42

## Se x for maior que 1...
if x > 1:
  ## Diga que ele é maior que 1
  print("É maior que 01.")
  ## Diga que ele é menor que 100
  if x < 100:
    print("Menor que 100")
````

### Alavancas If.. Else: Decisões em dois caminhos

Vimos que o "If" permite tomar decisões simples em processos. Lembrando do primeiro código que fizemos da seguinte forma:

![img02ifCode](https://i.imgur.com/gTpuxh8.png)

Porém aqui, nada irá acontecer caso `x` não seja exatamente igual a 2, e o código continua correndo como se nada tivesse acontecido...

![img04ifNothing](https://i.imgur.com/Noev8vH.png)

Acontece que muitas vezes queremos que nosso processo se comporte de outra forma, e tome decisões ativamente quando a condição não for verdadeira.

![img03ifElse](https://i.imgur.com/ktfMvUv.png)

Neste caso, isso pode ser feito com uma condição da seguinte forma:

```
SE x for 2, então:
  imprima: "É verdade"
DO CONTRÁRIO:
  imprima: "É mentira!"
```

O que na Python, é uma estrutura de código chamada "If... else"

````
## Declaramos um x
x = 5

## Vamos montar o nosso laço!
if x == 2:
  print("É verdade! X é igual a 2!")
else:
  print("É mentira! X é diferente de 2!")
````

## Um pouco de bagunça: funções com "if else"

Apenas um lembretezinho: podemos fazer funções de tudo, como se fossem pequenas caixinhas reaproveitáveis.

Uma função teria a seguinte estrutura:

![img05fn](https://i.imgur.com/jGl9sVt.png)

````
def eh_dois(arg):
  if arg == 2:
    print("É dois mesmo.")
  else:
    print("Não é dois.")

## Cinco é dois?
eh_dois(5)
````

- Assim como anteriormente, podemos encaixar "if else" dentro de outros "if else" (e encaixar qualquer coisa dentro de qualquer coisa na verdade!):

![img06scheme](https://i.imgur.com/K0R7KQL.png)

````
## Começamos com x
x = 5

## Se x for maior que 100, me diga que ele é menor que 50.
if x > 100:
  print("x é maior que 100.")
  ## Do contrário, me faça outros testes.
else:
  ## Se x for menor que 50, me diga que ele é menor que 50.
  if x > 50:
    print("x é maior que 50.")
    ## Do contrário faça outros testes.
  else:
      ## Se x for maior que 20, me diga que ele é.
      if x > 20:
        print("x é maior que 20")
        ## Do contrário, teste se ele é 2.
      else:
        eh_dois(x)
````

# Estruturas de controle: laços e o laço "para cada..."

Vimos que as estruturas do tipo "If" realizam operações somente caso passem em alguma condição de teste.

Ao contrário de "If", as vezes simplesmente se deseja repetir um processo um número de vezes, que pode ser pré-determinado ou não.

Suponhamos que nós temos a seguinte lista:

````
## Lista de valores
valores = [10, 20, 30, 50, 80, 100, 150]
````

Queremos testar se algum valor dessa lista é maior que 100. Podemos fazer da seguinte forma usando o que aprendemos anteriormente:

````
# O primeiro elemento de 'valores' é maior que 100?
## Lembre-se que na Python começa-se a contar do zero:
valores[0] > 100
````

````
# O segundo elemento de 'valores' é maior que 100?
valores[1] > 100
````

````
# O terceiro elemento de valores é maior que 100?
valores[2] > 100
````

Percemos rapidamente que o processo se torna entediante e improdutivo. Sabidamente, foi inventada uma forma de percorrer algo, e repetir processos quantas vezes fossem necessários: os laços `for`, ou, em português...

Em algumas situações é comum que uma mesma instrução (ou conjunto delas) precise ser executada várias vezes seguidas. Nesses casos, normalmente utilizamos um loop (ou laço de repetição) que permite executar o mesmo bloco de código enquanto uma condição é atendida. Em Python, os loops são codificados com as estruturas de repetição `for` e `while`.

```
PARA CADA X
  FAÇA processo
```

Vamos testar com o nosso exemplo proposto acima, testando se os valores são maiores que 100. 

Primeiro, vamos percorrer e exibir os valores um a um usando um `for` simples:

````
## Primeiro, vamos exibir os valores, um por um.
for valor in valores:
  print(valor)
````

Vamos fazer outro exemplo.
Enquanto percorremos a lista de valores, a variável indicada no `for` receberá, a cada iteração, um item da coleção. Assim, podemos executar algum processamento com esse elemento. No código abaixo percorremos a lista nomes e imprimimos cada elemento.

````
# Para a seguinte frase:
frase = "Ola Mundo"
# Iremos imprimir a frase
for x in frase:
    print(x)
````

````
nomes = ['Pedro', 'João', 'Leticia']
for n in nomes:
     print(n)
````

A variável definida na linha 1 é uma lista inicializada com uma sequência de valores do tipo string. A instrução for percorre todos esses elementos, um por vez e, em cada caso, atribui o valor do item à variável n, que é impressa em seguida. O resultado, então, é a impressão de todos os nomes contidos na lista.

Esse código opera da seguinte forma:

1. Primeiro, começa do valor inicial:

![img07firstValue](https://i.imgur.com/FMCwxDt.png)

2. A seguida, passa para o próximo valor:

![img08nextValue](https://i.imgur.com/Ql8tFU3.png)

3. E assim por diante, até o último valor:

![img09finalValue](https://i.imgur.com/TRAUUeT.png)

Assim, o processo completo (EXIBIR VALOR) irá rodar exatamente uma vez para cada um dos valores da nossa lista inicial.

Agora, vamos nos aproveitar da mesma estrutura, para decidir quais valores são maiores que 100, que era o problema original a qual perseguimos.

````
## Testa cada um dos valores, verificando se é maior que 100.
for valor in valores:
  if valor > 100:
    print("O valor", valor, "é maior que 100.")
````

Note que no exemplo acima, misturamos duas estruturas de controle de processo: um "if", que já haviamos aprendido, e um "for", novidade, que visualmente são resumidos:

![img10structure](https://i.imgur.com/JruqWzP.png)

- Como tudo em programação, pode-se colocar um "for" dentro de um "for"! Veja só:


````
## Vamos criar alguma variáveis.
letras = "abcde"
números = [1, 2, 3]

## Para cada letra em letras;
  ## Para cada número em números;
  ## Imprima um número de vezes a mesma letra
for letra in letras:
  for número in números:
    print(letra*número)
````

Cuja estrutura visual é da forma:

![img12twoFors](https://i.imgur.com/C78Xeql.png)

Outra combinação que podemos fazer é `for` com `else` como feito com os exemplos do `If`. 
É possível adicionar a instrução else ao final do for. Isso faz com que um bloco de código seja executado ao final da iteração, como mostra o exemplo a seguir:

````
nomes = ['Pedro', 'João', 'Leticia']
for n in nomes:
     print(n)
else:
     print("Todos os nomes foram listados com sucesso")
````

Na primeira linha definimos uma variável que armazenará uma lista de nomes. Após isso, a instrução `for` percorre todos esses elementos e atribui um a um à variável n, que será impressa, como pode ser visto na linha 3. Após o loop se encerrar, o bloco de código contido na instrução `else` é acionado, imprimindo a mensagem na tela.

## Condicional while

O comando `while` faz com que um conjunto de instruções seja executado enquanto uma condição é atendida. Quando o resultado dessa condição passa a ser falso, a execução do loop é interrompida, como mostra o exemplo a seguir:

O condicional "While" nada mais é que uma combinação de um laço tipo "for", mas delimitado por um operador booleano, que irá dizer quantas vezes ele irá rodar.

````
contador = 0
while (contador < 5):
       print(contador)
       contador = contador + 1
````

Neste código, enquanto a variável `contador`, inicializada com 0, for menor do que 5, as instruções das linhas 3 e 4 serão executadas.

Observe que na linha 4 incrementamos o valor da variável `contador`, de forma que em algum momento seu valor igualará o número 5. Quando isso for verificado na linha 2, o laço será interrompido. Sem esse código, a condição de parada não será atingida, gerando o que é chamado de loop infinito. Evite que isso aconteça, pois leva ao congelamento e finalização da aplicação.

````
# Começamos com 0 macaquinhos
macacos = 5

# Enquanto não tiverem 10 macaquinhos, vou dizer quantos há.
while macacos < 10:
  macacos = macacos + 1
  print("Tenho", macacos, "macacos.")
````

Assim como o `if` e `for` o `while`também pode ser usado com o `else`. Nesse caso ao final do `while` podemos utilizar a instrução `else`. O propósito disso é executar alguma instrução ou bloco de código ao final do loop, como podemos ver no exemplo a seguir:

````
contador = 0
while (contador < 5):
      print(contador)
      contador = contador + 1
else:
      print("O loop while foi encerrado com sucesso!")
````

Assim como acontece com `for/else`, declarando o `else` ao final do `while` é possível executar um código ao final do loop. Neste caso será impressa a mensagem: “O loop while foi encerrado com sucesso!”.

No loop `while`, a expressão é testada enquanto for verdadeira. A partir do momento que ela se torna falsa, o código da cláusula else será executado, se estiver presente.

````
x = 0
while x < 10:
    print(x)
    x += 1
else:
    print("fim while")
````

Como tudo em código, há muitas chances de tudo dar muito errado. Existem os infames "laços infinitos", veja:

````
## Enquanto verdade, imprima "Nada"
while True:
  print("Nada")
````

## Condicional Break

O comando `break`, sai imediatamente do laço de repetição mais interno, seja `for` ou `while`.
Se dentro da repetição for executado um break, o loop será encerrado sem executar o conjunto da cláusula else.

````
x = 0
while x < 10:
    print(x)
    x += 1
    if x == 6:
        print("x é igual a 6")
        break
else:
    print("fim while")
````
